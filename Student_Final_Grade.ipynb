# **STUDENT FINAL GRADE BASED ON DIFFERENT ATTRIBUTES**

## Loading the Libraries

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as pyplot
from matplotlib import style
```

## Load the Data

```python

df = pd.read_csv("student-mat.csv", sep=";")
```

## Selecting some Specific Attributes and implementing the Model

```python
# Features and labels
df = df[['G1', 'G2', 'G3','traveltime','studytime', 'failures', 'absences']]
predict = "G3"
x = np.array(df.drop([predict], axis=1))
y = np.array(df[predict])

# Split the data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.1)

# Train the model
linear = LinearRegression()
linear.fit(x_train, y_train)

# Get the accuracy
acc = linear.score(x_test, y_test)
print(acc)

# Make predictions
predictions = linear.predict(x_test)
for x in range(len(predictions)):
    print(predictions[x], x_test[x], y_test[x])
```

## Plotting the Data

```python
p="G1"
pyplot.scatter(df[p],df["G3"])
pyplot.xlabel(p)
pyplot.ylabel("Final Grade")
pyplot.show()

q="G2"
style.use("ggplot")
pyplot.scatter(df[q],df["G3"])
pyplot.xlabel(q)
pyplot.ylabel("Final Grade")
pyplot.show()

r="absences"
style.use("ggplot")
pyplot.scatter(df[r],df["G3"])
pyplot.xlabel(r)
pyplot.ylabel("Final Grade")
pyplot.show()

s="studytime"
style.use("ggplot")
pyplot.scatter(df[s],df["G3"])
pyplot.xlabel(s)
pyplot.ylabel("Final Grade")
pyplot.show()
```
